```python
## Step 1: Import necessary libraries and load the IMDB dataset and load the model architecture summary
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import load_model
```

```python
# Load the IMDB dataset word index and create a reverse mapping of word indices to words
word_index = imdb.get_word_index()
reverse_word_index = {value: key for key, value in word_index.items()}
```

```python
# Load the pre-trained model with ReLU activation function
model = load_model('simple_rnn_model.h5')
model.summary() # Print the model architecture summary
```

```python
model.get_weights() # Get the weights of the model layers
```

```python
# Step 2: Helper function to decode the review from word indices back to text
def decode_review(encoded_review):
    return ' '.join([reverse_word_index.get(i - 3, '?') for i in encoded_review])

# Function to preprocess the input review text
def preprocess_review(review_text):
    # Tokenize the review text into words
    words = review_text.lower().split()

    # Convert words to their corresponding indices based on the word index
    encoded_review = [word_index.get(word, 2) + 3 for word in words]  # Add 3 to account for reserved indices

    # Pad the encoded review to a fixed length (e.g., 500)
    padded_review = sequence.pad_sequences([encoded_review], maxlen=500)
    return padded_review
```

```python
# Prediction function to classify the sentiment of a review
def predict_review_sentiment(review_text):
    # Preprocess the review text
    processed_review = preprocess_review(review_text)

    # Make a prediction using the loaded model
    prediction = model.predict(processed_review)

    # Interpret the prediction (0 for negative, 1 for positive)
    sentiment = 'Positive' if prediction[0][0] >= 0.5 else 'Negative'
    return sentiment, prediction[0][0]
```

```python
# Step 4: User input for review text and prediction
# Example usage
example_review = "This movie was fantastic! The acting was great and the story was thrilling."

sentiment, confidence = predict_review_sentiment(example_review)

print(f"Review: {example_review}")
print(f"Sentiment: {sentiment}")
print(f'Prediction Confidence: {confidence:.4f}')
```
